```python
# Tuple, sıralı (ordered) ve değiştirilemez (immutable)  --> Farklı veri tipleri içerebilir, Hızlıdır, Tuple içinde tuple olabilir 

t0 = ()  # Boş tuple
t1 = tuple()  # Boş tuple
print(t0, t1) 

t2 = (5)      # Integer
t3 = (5,)     
t4 = "hello", 

print(type(t2)) 
print(type(t3))  
print(type(t4))  

# Slicing
t = (10, 20, 30, 40, 50, 60)

print(t[1:4])     
print(t[::-1])  # Ters çevirme
```

```python
# deque (Double-Ended Queue) --> Fifo,
# Collections modülünün bir parçası, Çift uçlu, Hızlı, Dairesel kullanım, thread-safe

from collections import deque

dq0 = deque()
print(dq0)

dq1 = deque([1, 2, 3])
dq1.append(4)  
dq1.appendleft(0)
print(dq1) 
dq1.pop() 
print(dq1) 
dq1.popleft()
print(dq1) 

dq = deque([1, 2, 3], maxlen=3) #Listenlerdekinin aksine boyutlandırma var.
print(dq)
dq.append(4)  # Yeni eleman eklenince en eski eleman (1) silinir
print(dq) 

#Listelerdeki fonk. hep fifo mantığında bunda kullanılıyor.
```

```python
# Resimlerde önemli kavramlar:

# Generator --> iteratör türü, gerektiği zaman veriyi üretir ve döndürür

# yield ifadesi kullanıldığı için fonksiyon generator haline gelir.


def createGenerator():
    liste = range(1, 4) 
    for i in liste:
        yield i  # i'yi döndürür ve fonksiyonu duraklatır

generator = createGenerator()  
print(generator)  # referans

for i in generator:  # Generator'un ürettiği değerler üzerinde döngü kurar
    print(i)  # Her bir değeri ekrana yazdırır

# yield sayesinde değerleri adım adım üretip sonsuz falan yapabilirim. --> bu iterasyon iste ya
```

```python
# OS (Operating System) modülü ile çalışma
import os  # İşletim sistemi işlemleri için os modülü dahil edilir.

# İşletim sisteminin türünü ekrana yazdırır (Windows için 'nt', Linux/macOS için 'posix' döner)
print(os.name)

currentDir = os.getcwd()
print("Şu anki çalışma dizini:", currentDir) # current working directory - cwd

# Yeni Klasör (Folder) Oluşturma
folder_name = "new_folder"  # Oluşturulacak klasörün adı belirlenir
os.mkdir(folder_name)  # Belirtilen isimde yeni bir klasör oluşturulur

# Klasörün Adını Değiştirme
new_folder_name = "new_folder_2"  # Klasörün yeni adı belirlenir
os.rename(folder_name, new_folder_name)  # Klasörün adı değiştirilir

#  Çalışma Dizinini Değiştirme
# Yeni oluşturulan klasöre geçiş yapılır
os.chdir(currentDir + "\\" + new_folder_name)
print("Yeni çalışma dizini:", os.getcwd())  # Yeni çalışma dizini ekrana yazdırılır

# Eski çalışma dizinine geri dönme
os.chdir(currentDir)
print("Geri dönülen çalışma dizini:", os.getcwd())

#  Mevcut Dizin İçindeki Dosyaları Listeleme
files = os.listdir()  # Geçerli dizindeki dosyaları listeler
print("Mevcut dizindeki dosyalar:", files)

# Belirli Uzantıya Sahip Dosyaları Filtreleme
for f in files:  # Dizindeki tüm dosyalar arasında dolaş
    if f.endswith(".py"):  # Eğer dosyanın uzantısı '.py' ise
        print("Python dosyası bulundu:", f)  # Dosya adı ekrana yazdırılır

#  Klasörü Silme
os.rmdir(new_folder_name)  # Daha önce oluşturulan klasör silinir

#  Tüm Dizinleri ve Dosyaları Gezme
# os.walk(), belirtilen dizindeki tüm dosya ve klasörleri listeler
for i in os.walk(currentDir):  
    print("Klasör içeriği:", i)

#  Dosyanın Var Olup Olmadığını Kontrol Etme
# Belirtilen dosyanın mevcut olup olmadığını kontrol eder
file_exists = os.path.exists("python_hatirlatma.py")
print("Dosya mevcut mu?", file_exists)  # True veya False döner
```
